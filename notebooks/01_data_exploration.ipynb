Generate the synthetic regime-switching market and inspect data quality.

```python
import matplotlib.pyplot as plt

from alphaforge.data import SyntheticMarketConfig, data_quality_report, generate_synthetic_market

panel = generate_synthetic_market(SyntheticMarketConfig(n_symbols=10, n_days=1500, seed=42))
panel.head()
```

```python
quality = data_quality_report(panel)
quality
```

```python
bench = panel[panel['symbol'] == 'BENCH'].set_index('date')
fig, axes = plt.subplots(2, 1, figsize=(10, 6), sharex=True)
axes[0].plot(bench.index, bench['close'])
axes[0].set_title('Benchmark close (synthetic)')
ret = bench['close'].pct_change()
axes[1].plot(ret.index, ret.rolling(20).std() * 252**0.5)
axes[1].set_title('Trailing 20d vol (annualized)')
plt.tight_layout()
```

Volatility clustering from the 2-state Markov regime is clearly visible — this is what the HMM regime model estimates causally in notebook 02.
