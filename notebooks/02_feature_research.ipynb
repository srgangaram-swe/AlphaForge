> **Educational research only.** Synthetic data, simulated results, not financial advice.

Build the causal feature panel and inspect the HMM stress-probability feature.

```python
import matplotlib.pyplot as plt

from alphaforge.data import SyntheticMarketConfig, generate_synthetic_market
from alphaforge.features import build_features, feature_columns

panel = generate_synthetic_market(SyntheticMarketConfig(n_symbols=10, n_days=1500, seed=42))
features = build_features(panel, benchmark_symbol='BENCH')
print(len(feature_columns(features)), 'features')
features.tail(3)
```

```python
regime = features.groupby('date')[['bench_vol', 'hmm_stress_prob']].first().dropna()
fig, ax1 = plt.subplots(figsize=(10, 4))
ax1.plot(regime.index, regime['bench_vol'], color='tab:gray', label='bench vol')
ax2 = ax1.twinx()
ax2.plot(regime.index, regime['hmm_stress_prob'], color='tab:red', alpha=0.7, label='HMM stress prob')
ax1.set_title('Causal HMM stress probability vs realized benchmark vol')
plt.tight_layout()
```

```python
corr = features[feature_columns(features)].corr()
plt.figure(figsize=(8, 7))
plt.imshow(corr, cmap='RdBu_r', vmin=-1, vmax=1)
plt.colorbar()
plt.title('Feature correlation')
plt.tight_layout()
```

The stress probability is built with expanding refits + *filtered* inference — tests/test_leakage.py proves values at *t* are unchanged when future data is mutated.
